```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
items = os.listdir('opensea_scrapings')

csv_files = []
for names in items:
    if names.endswith(".csv"):
        csv_files.append('opensea_scrapings/' + names)

df = pd.concat([pd.read_csv(f) for f in csv_files])
```

```python
## Extracting numbers from strings 

def get_number(seq):
    seq_type= type(seq)
    return seq_type().join(filter(seq_type.isdigit, seq))


def process_string(string):
    numbers_list = [get_number(f) for f in string.split('.')]
    numbers_sizes = [len(i) for i in numbers_list]
    ix = list(range(0, np.argmax(numbers_sizes)+1))

    list_to_join = [str(j) for i,j in enumerate(numbers_list) if i in ix]  # filtering when 
    return '.'.join(list_to_join)


df['price_clean'] = [process_string(f) for f in df.price]
```

```python
## drop rows which won't become floats
df = df[ pd.array([len(f) for f in df.price_clean]) > 0 ]
df['price_clean'] = df.price_clean.astype(float)
```

```python
plt.hist(df.price_clean, bins = 100)
plt.show()
```
